# 3. Variables

You can think of the variable as a name you give to a value in order to store it. Each variable is stored in a particular memory location based on its data type. Once you create the variable, you can access and manipulate it later.

What is actually happening behind the scene: when we assign the value to the variable, Python gives the reference of value (object) to the variable. Technically, the name of variable we chose is called Python `identifier`.

## 3.1 Declaring a variable

Declaring a variable is done by simply assigning a value to a variable name you choose by using assignment operator: **=**. 

Let´s store the sentence 'Have a great day!' to text variable:

````
# Declaring a variable
text = 'Have a great day!'    # Text is assigned to text variable
````

Once we create a variable we can print the output using `print()` function. Or we just type variable name in the code cell.

````
# Printing text variable
print(text)
````

 
Let´s create result_1 variable that holds number 123 and print the output. 


````
# Declaring a variabe
result_1 = 123
````

````
# Printing the output
print(result_1)
````

> Note: 
> Remember that assignment syntax must be as following. Firstly, you define a variable name that is to the left of the assignment operator. Then write a value to the right of operator.


```
my_variable = 224
```
NOT oppositely:


```
224 = my_variable
```




Once we declare a variable we can substitute its value by reassigning. The original result_1 variable holds number 123. Below we are assigning the new value 1500.

````
result_1 = 1500
print(result_1)
````

We can also assign some mathematical operation to a variable as you can see in the below example. When we call `print()` and pass result_2 as argument, the result of this operation will be returned.

````
# Declaring a variabe with mathematical operation
result_2 = (15**2) - (28//2)
# Printing the output
print(result_2)
````

Since result_2 variable is stored in the specific storage location we can access it and use to create a new variable. Let´s say that we want to substract number 11 from result_2 and store the output in a new variable. We can do following:

````
# Creating a new variable using the output of result_2
result_2_updated = result_2 - 11
# Printing the output
print(result_2_updated)
````

>For curiosity: As we already mentioned, each variable is actually an identifier in Python. When declaring, this identifier is given a unique number which is simply said, the exact place in memory, where variable "lives". You can create as many variables as you want and still no 2 variables will have the same unique number. You can return variable´s unique number using `id()` function.

````
print('Variable result_2: ',id(result_2),'\n','Variable result_2_updated: ',id(result_2_updated))
````

Let´s store another numeric type float - floating point number which is characteristic with a decimal point.

````
# Declaring a variable
my_float = 148.84
# Printing the output
print(my_float)
# Returning the data type
print(type(my_float))
````

## 3.3 Naming conventions and some rules 

> Let´s take an example (although maybe it is not so realistic). Suppose you are taking aside bills from your shopping in the supermarket for the period of almost 6 months. You are recording purchase date, information about the price, category and the quantity of each item you bought, the name of item, etc. Now, you would like to summarize how much money you spend on the specific category. There could be groceries, cosmetics, games, restaurants, food articles used to make a nice dinner for your friends, etc.

> You want to focus on a specific category - games category - because you are talking to yourself when looking at your collection: "Hm this is great, I received all possible achievements in the games I recently purchased! Really, practice makes perfect! I also purchased many items I always wanted in my favorite games that are so cool. Wait...how much money did I actually spend? 

You decided to filter your data only for games category and make some visualization or summary statistics about your spending over time. The first step is create a variable in order to store the data you want to analyze. 

### 3.3.1. Variable names must start with a letter / underscore

There are few rules that we need follow in order to create a proper variable name and avoid unnecessary error messages. Let´s think about possible variable names for data about gaming:

a) `games` / `Games` 

- variable names can contain lowercase and uppercase letters as well
- special symbols such as %, * or $ are not allowed

b) `_games` 
- variable names can start with the underscore

c) `games_2021`   
- you can also use digits

> IMPORTANT: Although variable names can consist of digits from 0-9, they can´t start with any digit. That would throw a SyntaxError message.

### 3.3.2 Variable names are case sensitive

Python distunguishes between lovercase and uppercase letters. Thus if we create names `GameS` and `GAMES`, they are NOT the same variables.

### 3.3.3 Awareness of Python reserved words

There are few specific words that you can not use as the name of variable. They are called **keywords** and have a specific purpose and meaning in Python language. Whenever you would have doubt whether a particular word you want to choose is a keyword, you can check all of them using `help()` function:






````
# Getting all Python´s reserved words
help('keywords')
````

### 3.3.4 Naming conventions

- meaningful variable names

When creating a variable where we want to store the data about games, a possible name that could come to mind could be some abbreviation: 

`gs` - the first and the last letter of games for example

Yes looks fine, less typing. But after dozens of code lines, integrating this variable into bigger code block and coming back to your code after some time you easily can forget what this variable is supposed to mean. 

Okay, we can create a variable name that more describes what data it is holding. When we come back to our script even after few weeks, we would definitely understand the code.










- readable variable names

Even if we created a comprehensive variable name, that describes the stored data, it might happen that it will become a mess. For example, `gamescategorypurchasesinapril` variable name tells us that we stored the data about purchases in games category throughout the april. Although we surely can write it this way, it is really hard to read not only for you, but also for others who might read the code you wrote. 
var_name.jpg  





If you want to write a good descriptive variable name, the better option would be to write it using so-called snake case where you separate words with the undescore. And choose more succint words:

`games_purchases_april`

Note that this part conceives variable name conventions, not rules, thus you can use whichever style you want in order to create a name. On the other hand keep in mind that readibility really matters.
